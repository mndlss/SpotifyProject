# Cluster Graph

```python
y=df.iloc[:,11:23]


from sklearn.cluster import KMeans
km=KMeans(n_clusters=10, n_init=10) #because 10 coloumns for cluster graph
km.fit(y)


fig,axes=plt.subplots()


axes.scatter(y.danceability, y.energy, y.key)
axes.scatter(y.loudness, y.speechiness, y.acousticness)
axes.scatter(y.instrumentalness, y.liveness)
axes.scatter(y.tempo, y.duration_ms)

plt.title('Cluster Graph')
plt.show()
```

```python
def time_series_multiline(df, track_album_release_date, danceability, playlist_genre):
    sns.set_palette("Dark2")
    
    if playlist_genre:
        for genre, series in df.groupby(playlist_genre):
            plt.plot(series[track_album_release_date], series[danceability], label=genre)
        plt.legend(title=playlist_genre, bbox_to_anchor=(1, 1), loc='upper left')
    else:
        plt.plot(df[track_album_release_date], df[danceability])

    sns.despine()
    plt.xlabel('track_album_release_date')
    plt.ylabel('danceability')
    plt.show()

time_series_multiline(df, 'track_album_release_date', 'danceability', 'playlist_genre')
```
